```python
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first='True')
dataset
```

```python
independent  = dataset[['R&D Spend', 'Administration', 'Marketing Spend',
        'State_Florida','State_New York']]
independent
```

```python
dependent = dataset[['Profit']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent, test_size = .30, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor =DecisionTreeRegressor(criterion='poisson',splitter='best')   #best   #random
regressor=regressor.fit(X_train,Y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree (regressor)
plt.show
```

```python
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score= r2_score(Y_test,Y_pred)
r_score
```
